```python
def am_pm(str):
    judge = ""
    flag = 0
    for chr in str:
        if chr==":":
            break
        if flag==1:
            judge+=chr
        if chr == " ":
            flag=1;
    return judge

        
```

```python
timee = df_2010["InvoiceDate"].map(lambda x:am_pm(x)) # for how much time the users were most active (18)

len(timee.groupby(timee).count())
```

```python
timee.groupby(timee).count().sort_values(ascending=0)
```

```python
# on which date most orders were placed (19df_)
def dts(str):
    day = ""
    flag = 0
    for chr in str:
        if flag==1 and (chr=='/' or chr=='-'):
            break
        if flag==1:
            day+=chr
        if chr == '/' or chr=='-':
            flag=1
            
    return day
```

```python
day = df_2010["InvoiceDate"].map(lambda x:dts(x)) # for how much time the users were most active
timee.groupby(day).count().sort_values(ascending=0)
```
